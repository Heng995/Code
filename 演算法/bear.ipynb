# 熊種類辨識

##### 辨識 北極熊 黑熊 大熊貓 棕熊

```python
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

# Load the model
model = load_model('keras_model.h5')
print("done")
```

```python
classes = ['北極熊', '黑熊', '大熊貓', '棕熊']
```

```python
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
image = Image.open('Panda01.jpeg')
plt.imshow(image)
```

```python
size = (224, 224)
image = ImageOps.fit(image, size, Image.ANTIALIAS)
plt.imshow(image)
```

```python
image_array = np.asarray(image)
normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
data[0] = normalized_image_array
```

```python
prediction = model.predict(data)
print(prediction)
index = np.argmax(prediction)
print(index)
print(classes[index])
```

```python
bears = ['black_bear01.jpeg', 'brown_bear01.jpeg', 'panda01.jpeg', 'polar_bear01.jpeg', 'panda02.jpeg', 'polar_bear02.jpeg', 'brown_bear02.jpeg', ]
```

```python
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

for bear in bears:
  image = Image.open(bear)
  plt.imshow(image)
  size = (224, 224)
  image = ImageOps.fit(image, size, Image.ANTIALIAS)
  image_array = np.asarray(image)
  normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
  data[0] = normalized_image_array
  prediction = model.predict(data)
  print(prediction)
  index = np.argmax(prediction)
  print(classes[index])
```
